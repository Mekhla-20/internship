## Fetching from yatra.com

```python
url='https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=19%2F09%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=1235761094117'
driver.get(url)
```

```python
airline=[]
doj=[]
src=[]
dst=[]
stops=[]
duration=[]
dept_time=[]
arrival_time=[]
price=[]
```

```python
air_name=driver.find_elements(By.XPATH,'//div[@class = "flight-det table full-width clearfix"]')       #scraping urls
for _ in range(60):
    for i in air_name:
        if(i.text !=" "):
            print(i.text.split("\n"))
            airline.append(i.text.split("\n")[0])
            doj.append("19/09/22")
            src.append(i.text.split("\n")[3])
            dept_time.append(i.text.split("\n")[2])
            arrival_time.append(i.text.split("\n")[4])
        
            if i.text.split("\n")[5]=="+ 1 day":
                dst.append(i.text.split("\n")[6])
                stops.append(i.text.split("\n")[8])
                duration.append(i.text.split("\n")[7])
                price.append(i.text.split("\n")[9])
            
            else:
                dst.append(i.text.split("\n")[5])
                stops.append(i.text.split("\n")[7])
                duration.append(i.text.split("\n")[6])
                price.append(i.text.split("\n")[8])
            
    driver.execute_script("window.scrollTo(0,1000)")
    time.sleep(3)
```

```python
print(len(airline))
print(len(doj))
print(len(src))
print(len(dst))
print(len(stops))
print(len(duration))
print(len(dept_time))
print(len(arrival_time))
print(len(price))
```
