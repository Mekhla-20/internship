## Q2)In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv

```python
BrandName=[]
ProdName=[]
Price=[]
Delivery=[]
ProdURL=[]
for j in range(3):
    time.sleep(5)
    try:
        brand=driver.find_elements(By.XPATH,"//span[@class='a-size-base-plus a-color-base a-text-normal']")
        price=driver.find_elements(By.XPATH,"//span[@class='a-price-whole']")
        delivery=driver.find_elements(By.XPATH,"//span[@class='a-color-base a-text-bold']")
        produrl=driver.find_elements(By.XPATH,"//a[@class='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal']")
        
        for i in brand:
            BrandName.append(i.text)
            ProdName.append(product)
        for i in price:    
            Price.append(i.text)
            ProdName.append(product)
        for i in delivery:
            Delivery.append(i.text)
        for i in produrl:
            ProdURL.append(i.get_attribute('href'))
        
    
            
       
        driver.execute_script("window.scrollBy(0,1000)")
        next_button=driver.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        next_button.click()
    except NoSuchElementException:
        BrandName.append("Not present")

print("\n\nBrand Name:::\n",BrandName)   
print(len(BrandName))

print("\n\nPrice:::\n",Price)   
print(len(Price))

print("\n\nDelivery:::\n",Delivery)   
print(len(Delivery))

print("\n\nURL:::\n",ProdURL)   
print(len(ProdURL))


        
```

```python
Exchange=[]
Available=[]
for i in ProdURL[:20]:
    driver.get(i)
    time.sleep(5)
    
    try:
        avail=driver.find_elements(By.XPATH,"//span[@class='a-size-medium a-color-success']")
        ret=driver.find_elements(By.XPATH,"//div[@class='a-row icon-farm-wrapper']")
        for j in avail:
            Available.append(j.text)
        for j in ret:
            Exchange.append(j.text.split("\n")[1])
            
    except NoSuchElementException:
        Available.append("Not present")


print("\n\nAvailability:::\n",Available)   
print(len(Available))
print("\n\nExchange:::\n",Exchange)   
print(len(Exchange))
```

## Q3) Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
fruits=[]
cars=[]
machineLearning=[]
guitar=[]
cakes=[]
product=['fruits','cars','machine learning','guitar','cakes']
prod_list=[fruits,cars,machineLearning,guitar,cakes]

def images(item,li):
    url='https://images.google.com/'
    driver.get(url)
    search_loc=driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input")
    search_loc.send_keys(item)
    search_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button")
    search_button.click()
    imgs=driver.find_elements(By.XPATH,"//img[@class='rg_i Q4LuWd']")
    for image in imgs[0:10]:
        li.append(image.get_attribute('src'))
        
for i in range(0,5):
    images(product[i],prod_list[i])
    time.sleep(5)
    print("\n\n",prod_list[i])
```

## Q7) Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
url='https://www.digit.in/top-products/best-gaming-laptops-40.html'
driver.get(url)
```

```python
prod_name=[]
processor=[]
display=[]
os=[]
memory=[]
graphics_pro=[]
body=[]
price=[]
seller=[]

name=driver.find_elements(By.TAG_NAME,"h3")
for i in name:
    #print(i.text)
    prod_name.append(i.text)

details=driver.find_elements(By.XPATH,"//div[@class='Spcs-details']")
for i in details:
    processor.append(i.text.split("\n")[1].split(":")[1])
    display.append(i.text.split("\n")[2].split(":")[1])
    os.append(i.text.split("\n")[3].split(":")[1])
    memory.append(i.text.split("\n")[4].split(":")[1])
    graphics_pro.append(i.text.split("\n")[5].split(":")[1])
    body.append(i.text.split("\n")[6].split(":")[1])

pr=driver.find_elements(By.XPATH,"//td[@class='smprice']")
for i in pr:
    price.append(i.text)
    
sr=driver.find_elements(By.XPATH,"//td[@class='smmerchant']")
for i in sr:
    seller.append(i.text)

prod_details=pd.DataFrame(list(zip(prod_name,processor,display,os,memory,graphics_pro,body,price,seller)),columns=['Brand','Processor','Display','OS','Memory','Graphics Processor','Body','Price','Seller'])
prod_details
```

## Q10) Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price, facilities and property description

```python
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
url='https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-07-29&to=2022-08-01&guests=2&HostelNumber=&page=1'
driver.get(url)
```

```python
name=[]
dist=[]
rating=[]
reviews=[]
oreviews=[]
pprice=[]
dprice=[]


n=driver.find_elements(By.XPATH,"//h2[@class='title title-6']")
for i in n:
    name.append(i.text)

d=driver.find_elements(By.XPATH,"//span[@class='description']")
for i in d:
    dist.append(i.text.split("-")[1].split(" ")[1])

r=driver.find_elements(By.XPATH,"//div[@class='score orange big']")
for i in r:
    rating.append(i.text)

re=driver.find_elements(By.XPATH,"//div[@class='reviews']")
for i in re:
    reviews.append(i.text.split(" ")[0])
    
ore=driver.find_elements(By.XPATH,"//div[@class='keyword']")
for i in ore:
    oreviews.append(i.text.split(" ")[0]) 
    
pp=driver.find_elements(By.XPATH,"//a[@class='prices']")
for i in pp[3:]:
    print(i.text.split("\n"))
    if i.text.split("\n")[0]=='No Privates Available':
        pprice.append("No Privates Available")
        dprice.append(i.text.split("\n")[2])
    else:
        pprice.append(i.text.split("\n")[1])
        if i.text.split("\n")[2]=='No Dorms Available':
            dprice.append('No Dorms Available')
        else:
            dprice.append(i.text.split("\n")[3])
```

```python
facility=[]

f=driver.find_elements(By.XPATH,"//div[@class='rating-factors prop-card-tablet rating-factors small']")
for i in f:
    facility.append(i.text.replace("\n",",")) 
 
```

```python
description=[]
URL=[]

url=driver.find_elements(By.XPATH,"//a[@class='view-button']")
for i in url:
  
    URL.append(i.get_attribute("href")) 

print(URL)
```
