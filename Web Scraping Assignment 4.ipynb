## Q1) Scrape the details of most viewed videos on YouTube from Wikipedia.Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videosYou need to find following details:
## A) Rank
## B) Name
## C) Artist
## D) Upload date
## E) Views

```python
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
import requests
import urllib
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException,StaleElementReferenceException
```

```python
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
driver
```

```python
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
# Scraping rank
r=driver.find_elements(By.TAG_NAME,"tr")
for i in r[2:32]:
    rank.append(i.text.split("\n")[0].split(".")[0])
    name.append(i.text.split("\n")[0].split(".")[1].split("[")[0])
#scraping Artist
c=4
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    
    artist.append(a[c].text)
    c=c+6
#scraping views
c=5
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    
    views.append(a[c].text)
    c=c+6
#scraping upload_date
c=6
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    upload_date.append(a[c].text)
    c=c+6

details=pd.DataFrame(list(zip(rank,name,artist,upload_date,views)),
                          columns=['Rank','Name','Artist','Upload Date','Views'])
details    
```

```python
name=[]
artist=[]
last_week_rank=[]
peak_rank=[]
weeks=[]

a=driver.find_elements(By.XPATH,"//ul[@class='lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max']")

artist.append(a[0].text.split("\n")[1])
last_week_rank.append(a[0].text.split("\n")[2])
peak_rank.append(a[0].text.split("\n")[3])
weeks.append(a[0].text.split("\n")[4])
    



#Scraping name
a=driver.find_elements(By.TAG_NAME,"h3")
for i in a[1:]:
    name.append(i.text)
while ("" in name):
    name.remove("")    
 

 #Scraping artist
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only']")
for i in a:
    artist.append(i.text)
 

#Scraping last_week_rank
j=0
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in range(len(a)):
    last_week_rank.append(a[j].text)
    j=j+3
    if (len(a)-1)<=j:
        break


#Scraping peak_rank
j=1
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in range(len(a)):
    peak_rank.append(a[j].text)
    j=j+3
    if (len(a)-1)<=j:
        break


#Scraping weeks
j=2
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in range(len(a)):
    weeks.append(a[j].text)
    j=j+3
    if (len(a)-1)<=j:
        break
while "" in last_week_rank: 
    last_week_rank.remove("")
while "" in peak_rank: 
    peak_rank.remove("")
while "" in weeks: 
    weeks.remove("")

```

```python
song_details=pd.DataFrame(list(zip(name[:100],artist,last_week_rank,peak_rank,weeks)),
                          columns=['Song Name','Artist','Last Week Rank','Peak rank','Weeks on board'])
song_details 
```

## Q9)Scrape the details most watched tv series of all time from imdb.com.Url = https://www.imdb.com/list/ls095964455/
## You have to find the following details:
## A) Name
## B) Year span
## C) Genre
## D) Run time
## E) Ratings
## F) Votes

```python
driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
driver
```

```python
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
name=[]
year_span=[]
genre=[]
run_time=[]
ratings=[]
votes=[]
#Scraping Name 
a=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']")
for i in a:
    name.append(i.text.split(".")[1].split("(")[0])
    
#Scraping year of span
a=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
for i in a:
    year_span.append(i.text.split("(")[1].split(")")[0])
#Scraping genre
a=driver.find_elements(By.XPATH,"//span[@class='genre']")
for i in a:
    genre.append(i.text)
#Scraping run time
a=driver.find_elements(By.XPATH,"//span[@class='runtime']")
for i in a:
    run_time.append(i.text)
#Scraping rating
a=driver.find_elements(By.XPATH,"//span[@class='ipl-rating-star__rating']")
for i in a:
    if i.text=="\n" or i.text=="Rate" or i.text=="":
        continue
    ratings.append(i.text)  
```

```python
##Scraping votes
a=driver.find_elements(By.XPATH,"//p[@class='text-muted text-small']")

for i in range(2,len(a),3):
    votes.append(a[i].text.split(":")[1])
```

```python
show_details=pd.DataFrame(list(zip(name,year_span,genre,run_time,ratings,votes)),
                          columns=['Show Name','Year of Span','Genre','Run Time','Rating','Votes'])
show_details
```
