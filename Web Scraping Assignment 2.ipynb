## Q1) : Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python


url='https://www.naukri.com/'
driver.get(url)
#Enter ‚ÄúData Analyst‚Äù in ‚ÄúSkill, Designations, Companies‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field
search_job=driver.find_element(By.CLASS_NAME,"suggestor-input")
search_job.send_keys("Data Analyst")
search_loc=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div[3]/div/div/div[5]/div/div/div/input")
search_loc.send_keys("Bangalore")

#Then click the search button.
search_button=driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/div[3]/div/div/div[6]")
search_button.click()
```

```python
#Then scrape the data for the first 10 jobs results you get

#scraping job title
title_tags=driver.find_elements(By.XPATH,"//a[@class='title fw500 ellipsis']")
job_titles=[]
for i in range(10):
    job_titles.append(title_tags[i].text)
print("\n\nJob titles:::\n",job_titles)    

#scraping company names
company_tags=driver.find_elements(By.XPATH,"//a[@class='subTitle ellipsis fleft']")
company=[]
for i in range(10):
    company.append(company_tags[i].text)
print("\n\nCompany:::\n",company)  

#scraping experience
exp_tags=driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi experience']")
exp=[]
for i in range(10):
    exp.append(exp_tags[i].text)
print("\n\nExperience:::\n",exp)  

#scraping location
loc_tags=driver.find_elements(By.XPATH,"//li[@class='fleft grey-text br2 placeHolderLi location']")
location=[]
for i in range(10):
    location.append(loc_tags[i].text)
print("\n\nLocations:::\n",location) 

#Finally create a dataframe of the scraped data.
jobs=pd.DataFrame()
jobs['Job Title']=job_titles
jobs['Company']=company
jobs['Experience']=exp
jobs['Location']=location
jobs
```

## Q10): Write a python program to scrape the salary data for Data Scientist designation.You have to scrape Company name, Number of salaries, Average salary, Minsalary, Max Salary from https://www.ambitionbox.com/

```python


driver=webdriver.Chrome(r"C:\Users\user\Downloads\chromedriver_win32\chromedriver.exe")
driver
url='https://www.ambitionbox.com/'
driver.get(url)
```

```python
#clicking on salaries button
salary_button=driver.find_element(By.XPATH,"/html/body/div[1]/nav/nav/a[4]")
salary_button.click()
```

```python
#setting the value in input field
search_salary=driver.find_element(By.XPATH,"/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input")
search_salary.send_keys("Data Scientist")
#selecting suggestion
select_sug=driver.find_element(By.XPATH,"/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]")
select_sug.click()
```

```python
#scraping company names, experience, salary record
title_tags=driver.find_elements(By.XPATH,"//div[@class='company-info']")
job_titles=[]
salary_record=[]
exp=[]

for i in range(10):
    job_titles.append(title_tags[i].text.split("\n")[0])
    exp.append(title_tags[i].text.split("\n")[2].split("(")[0].split("e")[0])
    salary_record.append(title_tags[i].text.split("\n")[2].split("(")[1].split(")")[0])
print("\n\ncompany names:::\n",job_titles)    
print("\n\nexperience:::\n",exp) 
print("\n\nsalary record:::\n",salary_record) 
```

```python
#scraping average salary
avg_salary_tags=driver.find_elements(By.XPATH,"//p[@class='averageCtc']")
avg_salary=[]
for i in range(10):
    avg_salary.append(avg_salary_tags[i].text)
print("\n\naverage salary:::\n",avg_salary)    
```

```python
#scraping minimum and maximum salary
salary_tags=driver.find_elements(By.XPATH,"//div[@class='salary-values']")
min_salary=[]
max_salary=[]
for i in range(10):
    min_salary.append(salary_tags[i].text.split("\n")[0])
    max_salary.append(salary_tags[i].text.split("\n")[1])
print("\n\nMinimum salary:::\n",min_salary) 
print("\n\nMaximum salary:::\n",max_salary)
```

```python
#Finally create a dataframe of the scraped data.
salary=pd.DataFrame()
salary['Company name']=job_titles
salary['experience']=exp
salary['Salary record']=salary_record
salary['average salary']=avg_salary
salary['minimum salary']=min_salary
salary['maximum salary']=max_salary
salary
```
